```python
import pandas as pd

df = pd.read_csv('netflix_movies (1).csv')

df.head(20)
```

```python
df.isnull().any()
```

```python
df[df['duration'].isna()==1]
```

```python
df['duration'] = df['duration'].fillna(0)
```

```python
df[df['rating'].isna()==1]
```

```python
df['rating'] = df['rating'].fillna(0)
```

```python
df.duplicated()
```

```python
df[df['title'].duplicated()==1]
```

```python
df = df.drop_duplicates()
```

```python
df['director'] = df['director'].fillna('')
df['cast'] = df['cast'].fillna('')
df['country'] = df['country'].fillna('')
```

```python
df.dtypes
```

```python
df['date_added'] = df['date_added'].fillna(0)
```

```python
df.head(20)
```

```python
df['show_id'] = df['show_id'].str[1:]
```

```python
df.isnull().any()
```

```python
df.head(50)
```

```python
df['duration_minutes'] = df['duration'].apply(
    lambda x: int(str(x).split()[0]) if isinstance(x, str) and 'min' in x.lower() else 0
)
df['duration_seasons'] = df['duration'].apply(
    lambda x: int(str(x).split()[0]) if isinstance(x, str) and 'season' in x.lower() else 0
)
```

```python
df.head(10)
```

```python
df = df.drop(columns=['duration', 'description'])
```

```python
df.head(10)
```

```python
result = df.groupby('type').size().reset_index(name='type_count')

print(result)
```

```python
result = df.groupby('rating').size().reset_index(name='type_count')

print(result)
```

```python
replacements = {
    '66 min': 0,
    '74 min': 0,
    '84 min': 0,
}

df['rating'] = df['rating'].replace(replacements)
```

```python
print(df.isnull().sum())
```

```python
df = df[df['show_id'] != 0]
```
